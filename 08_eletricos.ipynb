## Ônibus elétricos

Nesse notebook é feita uma análise dos ônibus elétricos

Os dados foram disponibilizados pela SMT e incluídos manualmente na pasta "data" do projeto

```python
import pandas as pd
from utils.load_csv import load_csv
```

```python
gdf_final = load_csv('gdf_final.csv')

df_eletricos = load_csv('Ônibus elétricos a bateria - 15 09 2025 - última(POR LOTE).csv', skiprows=4)
```

```python
df_eletricos
```

```python
df_eletricos = df_eletricos.dropna(axis=1, how="all")

df_eletricos = df_eletricos.dropna(axis=0, how="all")
```

```python
print(gdf_final['id_onibus'].dtype)

print(df_eletricos['Prefixo'].dtype)
```

```python
gdf_final['id_onibus'] = gdf_final['id_onibus'].astype(str).str.strip()

df_eletricos['Prefixo'] = df_eletricos['Prefixo'].astype(str).str.strip()
```

```python
print("id_onibus únicos:", gdf_final['id_onibus'].nunique())

print("prefixo únicos:", df_eletricos['Prefixo'].nunique())
```

```python
ids_gdf = set(gdf_final['id_onibus'].unique())

ids_eletricos = set(df_eletricos['Prefixo'].unique())

intersecao = ids_gdf.intersection(ids_eletricos)

print("Interseção:", len(intersecao))
```
