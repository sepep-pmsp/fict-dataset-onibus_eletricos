# LINHA
Aleatorizando linhas de ônibus, só pra Ana ter o que usar no bd

```python
posicoes_diaria = pd.read_csv(manual_data_path('posicoes-onibus-diario_2025-12-31_silver.csv'))

linhas = posicoes_diaria['nome_linha'].unique()

onibus_diesel_dia31['random_linhas'] = [
    random.choice(linhas) for _ in range(
        len(onibus_diesel_dia31)
    )
]
onibus_diesel_dia31
```

# LINESTRING

```python
linestring_dia31 = pd.read_csv(manual_data_path('linestring_2025-12-31_silver.csv'))
```

```python
onibus_diesel_dia31['geometry'] = linestring_dia31['geometry']
```

# POP AFETADA

```python
from utils.load_csv import load_csv
pop_afetada = load_csv('gdf_pop_afetada.csv')



pop_afetada_mp = pop_afetada['mp_pop_afetada'].unique()
pop_afetada_nox = pop_afetada['nox_pop_afetada'].unique()

onibus_diesel_dia31['random_pop_afetada_mp'] = [
    random.choice(pop_afetada_mp) for _ in range(
        len(onibus_diesel_dia31)
    )
]

onibus_diesel_dia31['random_pop_afetada_nox'] = [
    random.choice(pop_afetada_nox) for _ in range(
        len(onibus_diesel_dia31)
    )
]
```
